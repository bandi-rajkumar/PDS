b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from Mileage, CC from Engine, bhp from Power, and lakh from
New_price). (4 points)

```python
dataset['Mileage'] = dataset['Mileage'].str.extract(r'(\d+\.?\d*)')[0].astype(float)
dataset['Engine'] = dataset['Engine'].str.replace(' CC', '', regex=False).astype(float)
dataset['Power'] = dataset['Power'].str.replace(' bhp', '', regex=False).astype(float)
```

```python
dataset['New_Price'] = pd.to_numeric(dataset['New_Price'].str.replace(' lakh', '', regex=False), errors='coerce')
```

```python
dataset[['Mileage', 'Engine', 'Power', 'New_Price']].head(15)
```

```python
# Save the updated dataset to a CSV file
dataset.to_csv('afterTask(b).csv', index=False)
```

C) Change the categorical variables (Fuel_Type and Transmission) into numerical one hot
encoded value. (4 points).

```python
fuel=dataset['Fuel_Type'].unique()
transmission=dataset['Transmission'].unique()
print(fuel)
print(transmission)
```

```python
dataset = pd.get_dummies(dataset, columns=['Fuel_Type', 'Transmission'], drop_first=True, dtype=int)
dataset.head(10)
```

```python
# Save the updated dataset to a CSV file
dataset.to_csv('encodedDataset.csv', index=False)
```

```python
# Define mileage range
ranges = [0, 15, 20, 25, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']

dataset['Mileage_Category'] = pd.cut(dataset['Mileage'], bins=ranges, labels=labels)
dataset[['Mileage','Mileage_Category']].head(10)
```

```python
# Save the updated dataset to a CSV file
dataset.to_csv('addedFeature.csv', index=False)
```

```python
cols = dataset[['Name', 'Year', 'Mileage', 'Price','Owner_Type','Price','Seats','Power']] #select operation
cols
```

```python
rows_filetered = dataset[(dataset['Owner_Type'] == 'First') & (dataset['Price'] > 25)]#Filtering cars that are 'First' owner type and have a price greater than 25
rows_filetered
```

```python
cols = dataset.rename(columns=lambda x: {'Power': 'Horsepower', 'Seats': 'Num_Seats'}.get(x, x))
cols
```

```python
cols['Horsepower'] = cols['Horsepower'] * 1.34102  # 1 kW = 1.34102 BHP
cols['Horsepower']
```

```python
cols = cols.sort_values(by=['Mileage', 'Year'], ascending=[False, True]) #Arrange by Mileage (descending) and Year (ascending)
cols[['Mileage','Year']]
```

```python
# Group by 'Owner_Type' and calculating  mean price and mileage
summarized_dataset = cols.groupby('Owner_Type').agg(
    Average_Price=('Price', 'mean'),
    Average_Mileage=('Mileage', 'mean')
).reset_index()
summarized_dataset
```

```python
# Save the updated dataset to a CSV file
cols.to_csv('Operations.csv', index=False)
```
