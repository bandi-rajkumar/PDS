```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import necessary libraries
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Define file path and load dataset
dataset_path = '/content/drive/My Drive/StudentsPerformance.csv'
data = pd.read_csv(dataset_path)

# Display the first 10 rows of the dataset
data.head(10)
```

```python
# Verify the presence of missing values
missing_values = data.isnull().sum()
print(missing_values)
```

## Data Visualizations

```python
# Create a histogram for Math Scores
plt.figure(figsize=(10, 6))
sns.histplot(data['math score'], bins=10, color='skyblue')

# Add title and labels
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.ylabel('Count')

# Save the plot
plt.savefig('/content/drive/My Drive/Pds_Visualizations_Q2/Visual1.png')

# Display the plot
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(10, 6))

# Compute the correlation matrix for score columns
score_correlation = data[['math score', 'reading score', 'writing score']].corr()

# Generate a heatmap with annotations
sns.heatmap(score_correlation, annot=True, cmap='coolwarm', fmt=".2f")

# Add title
plt.title('Correlation Matrix of Scores')

# Save the plot
plt.savefig('/content/drive/My Drive/Pds_Visualizations_Q2/Visual2.png')

# Display the heatmap
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(10, 6))

# Create a box plot for Math Scores categorized by Gender
sns.boxplot(x='gender', y='math score', data=data, hue='gender', palette='Set2')

# Add title and labels
plt.title('Math Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')

# Save the figure
plt.savefig('/content/drive/My Drive/Pds_Visualizations_Q2/Visual3.png')

# Display the plot
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(10, 6))

# Create a scatter plot for Reading Scores vs. Writing Scores
sns.scatterplot(
    x='reading score',
    y='writing score',
    data=data,
    hue='gender',
    style='gender',
    palette='Set1'
)

# Add title and labels
plt.title('Relationship Between Reading and Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')

# Save the figure
plt.savefig('/content/drive/My Drive/Pds_Visualizations_Q2/Visual4.png')

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Compute average scores for each parental education level
average_scores = data.groupby('parental level of education')[
    ['math score', 'reading score', 'writing score']
].mean().reset_index()

# Set the index to parental education level
average_scores.set_index('parental level of education', inplace=True)

# Plot the average scores as a line plot
average_scores.plot(kind='line', marker='o', figsize=(10, 6))

# Add title and labels
plt.title('Average Scores Across Parental Education Levels')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')

# Save the figure
plt.savefig('/content/drive/My Drive/Pds_Visualizations_Q2/Visual5.png')

# Display the plot
plt.show()
```
